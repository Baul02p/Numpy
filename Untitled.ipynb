```python
# Array initialisation method
```

```python
#zeros array
zero_arr = np.zeros((2,3))
print(zero_arr)
```

```python
#one array
one_arr = np.ones((2,4))
print(one_arr)
```

```python
#full array
full_arr = np.full((3,2),8)
print(full_arr)
```

```python
#Identity Matrix  #Fill 1â€™s only on the diagonal (top-left to bottom-right)
id_arr = np.eye(2,2)
print(id_arr)
```

```python
#empty array   #np.empty(6) creates an array of 6 elements with random garbage values (not zeros).
print(np.empty(3))
```

```python
#evenly space array
print(np.arange(2,7,2))
```

```python
#specific number of values between a range
print(np.linspace(1,10,3))  #it means divide 1 to 10 number in 3 parts
```

```python
#random values array (float)
r_arr = np.random.rand(3,2)   #3,2 is row & col
print(r_arr)
```

```python
#random values array (int)
rint_arr = np.random.randint(1,10,(3,3))   #3,3 is row & col and 1,10 menas to print any no from 1 to 10
print(rint_arr)
```
