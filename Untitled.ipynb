```python
#Array Reshaping & Flattening
```

```python
arr1 = np.array([1,2,3,4,5,6])
print(arr1)
```

```python
reshaped = arr1.reshape([6,1])
print(reshaped)
```

```python
reshaped2 = arr1.reshape((2,3))
print(reshaped2)
```

```python
print(reshaped2.flatten())
```

```python
#Array Stacking & Splitting
```

```python
a = np.array([1,2,3])
b = np.array([4,5,6])

print(np.vstack((a,b)))  # vertical stacking - row wise
print(np.hstack((a,b)))  #horizontal stacking - col wise
```

```python
#Splitting
c = np.array([[1,2,3],[4,5,6]])
print(c)
```

```python
print(np.hsplit(c,3))
```

```python
print(np.vsplit(c,2))
```

```python
# Mathematical Operations on Multiple Arrays
```

```python
a = np.array([1,2,3])
b = np.array([4,5,6])
print(a)
print(b)
```

```python
print(np.add(a,b))
print(np.subtract(a,b))
print(np.multiply(a,b))
print(np.divide(a,b))
```

```python
print(np.dot(a,b))      # 4+10+18 = 32
```
