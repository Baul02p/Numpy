```python
import numpy as np #alias - np
```

```python
# 1. CREATING ARRAYS

list1 = [1,2,3]
print(list1)
```

```python
#1 dimensional array
arr_list1 = np.array(list1)
print(arr_list1)
```

```python
print(type(arr_list1 ))
```

```python
#2 dimensional array
list1 = [1,2,3]
list2 = [4,5,6]
arr_list2 = np.array([list1,list2])
print(arr_list2)
arr_list2.ndim
```

```python
a = np.array([[]])  #example
a.ndim
```

```python
#3 dimensional array
list1 = [1,2,3]
list2 = [4,5,6]
list3 = [7,8,9]
arr_list3 = np.array([[list1,list2,list3]])
print(arr_list3)
arr_list3.ndim
```

```python
arr1 = np.array([[1,2,3],[4,5,6]])
print(arr1)
```
